```python
import pandas as pd
```

```python
data=pd.read_csv("main.csv")
```

```python
l=[]
for i in range(len(data)):
    if data.loc[i]['COUNTRY'][0:3]=='USA':
        l.append(data.loc[i])
```

```python
df=pd.DataFrame(l)
```

```python
df.to_csv('filteredCountry.csv')
```

```python
id1=[]
for i in df['SKU']:
    if i not in id1:
        id1.append(i)
```

```python
f=['SKU','FIRST_MINIMUM_PRICE','SECOND_MINIMUM_PRICE']
final=pd.DataFrame(columns=f)
```

```python
for i in id1:
    z=[]
    fm=None
    sm=None
    for j in df[df['SKU']==i]['PRICE']:
        z.append(float(j[1:]))
    z.sort()
    if len(z)==1:
        fm=z[0]
        sm='Nan'
    elif len(z)==0:
        fm=0
        sm=0
    else:
        fm=z[0]
        sm=z[1]
    d={'SKU':i,'FIRST_MINIMUM_PRICE':fm,'SECOND_MINIMUM_PRICE':sm}
    final=final.append(d,ignore_index=True)
```

```python
final.to_csv("lowestPrice.csv")
```
